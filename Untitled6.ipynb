```python
mm=merged_matches
```

```python
mm.rename(columns={'id':'match_id'},inplace=True)
```

```python
mm['team1']=mm['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
mm['team2']=mm['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
```

```python
teams_to_remove = ['Rising Pune Supergiants','Rising Pune Supergiant','Kochi Tuskers Kerala','Pune Warriors','Gujarat Lions']

# Remove rows where 'team2' is in the list
mm = mm[~mm['team2'].isin(teams_to_remove)]
```

```python
teams_to_remove = ['Rising Pune Supergiants','Rising Pune Supergiant','Kochi Tuskers Kerala','Pune Warriors','Gujarat Lions']

# Remove rows where 'team1' is in the list
mm = mm[~mm['team1'].isin(teams_to_remove)]
```

```python
mm['team1'].value_counts()
```

```python
mm=mm[['match_id','city','winner','target']]
```

```python
delivery=delivery.merge(mm,on='match_id')
```

```python
#assigning small variable name for comfort
dl=delivery
```

```python
dl=dl[dl['inning']==2]
```

```python
dl['current_score']=dl.groupby('match_id')['total_runs'].cumsum()
```

```python
dl['runs_left']=dl['target']-dl['current_score']
```

```python
dl=dl[dl['runs_left']>-6]
```

```python
dl['balls_left']=126 - (dl['over']*6 + dl['ball'])
```

```python
dl
```

```python
dl['rrr']=dl['runs_left']/(dl['balls_left']/6)
```

```python
dl['crr']=dl['current_score']/((120-dl['balls_left'])/6)
```

```python
dl['result']='0'
```

```python
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
dl['result']=dl.apply(result,axis=1)
```

```python
dl['player_dismissed']=dl['player_dismissed'].fillna('0')
```

```python
dl['player_dismissed']=dl['player_dismissed'].apply(lambda x: x if x=='0' else 1)
```

```python
dl['player_dismissed']=dl['player_dismissed'].astype('int')
```

```python
wickets=dl.groupby('match_id')['player_dismissed'].cumsum()
```

```python
dl['wickets_left']=10-wickets
```

```python
pro=dl[['batting_team','bowling_team','city','target','current_score','runs_left','balls_left','wickets_left','rrr','crr','result']]
```

```python
pro.dropna(inplace=True)
```

```python
pro.info()
```

```python
pro.describe()
```

```python
pro=pro[pro['balls_left']!=0]
```

```python
pro=pro.sample(pro.shape[0])
```

```python
pro
```

TRAINING THE DATA

```python
from sklearn.model_selection import train_test_split
```

```python
x=pro.drop('result',axis=1)
y=pro['result']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,shuffle=False,test_size=.25)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(
    transformers=[
        ('ttf1', OneHotEncoder(sparse_output=False, drop='first'), ['batting_team', 'bowling_team', 'city'])
    ],
    remainder='passthrough'
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe=Pipeline([
    ('step1',ct),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
#Checking the data before fitting the pipeline
xtrain.isnull().sum()
```

```python
xtest.isnull().sum()
```

```python
pipe.fit(xtrain,ytrain)
```

```python
prediction=pipe.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(ytest,prediction)
```

```python
print('Accuracy of our model is:',accuracy)
```

```python
pipe.predict_proba(xtest)[10]
```
